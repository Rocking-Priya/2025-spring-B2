# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

If the viewer wants to examine the last three months of data first, we should focus on visual properties that immediately grab attention without requiring cognitive effort. Here are several ways you can achieve this:

1. Color Emphasis:

Use a brighter color for the last three months of data while keeping the earlier data in a muted or grayscale tone. The viewer’s eyes will naturally focus on the red-highlighted recent data.

2. Line Thickness & Style:

Make the last three months bolder (thicker line) while keeping the previous months in a thinner or dashed line. The bold, solid section stands out.

3. Annotating the Last Three Months:

Add markers or labels to highlight specific data points. The viewer’s attention will be drawn to the marked points.

```python
# 1. Color Emphasis:

ax = sns.lineplot(x=datearray[:-3], y=time_series[:-3], color="lightgray", alpha=0.5)
ax = sns.lineplot(x=datearray[-3:], y=time_series[-3:], color="red", linewidth=2)
```

```python
# 2. Line Thickness & Style:

ax = sns.lineplot(x=datearray[:-3], y=time_series[:-3], linestyle="dashed", color="gray")
ax = sns.lineplot(x=datearray[-3:], y=time_series[-3:], linestyle="solid", color="blue", linewidth=3)
```

```python
# 3. Annotating the Last Three Months:

ax = sns.lineplot(x=datearray, y=time_series, color="blue")
ax.scatter(datearray[-3:], time_series[-3:], color="red", s=100, label="Recent Data")
```

```python
import pandas as pd

# Creating the dataset
data = {
    "Country": ["AUS", "BRA", "CAN", "CHI", "FRA", "GER", "IND", "ITA", "MEX", "RUS", "SPA", "TUR", "UK", "US"],
    "A": [1, 1, 2, 1, 3, 3, 4, 2, 1, 4, 2, 7, 1, 1],
    "B": [2, 3, 3, 2, 2, 1, 1, 4, 5, 3, 3, 2, 2, 2],
    "C": [3, 4, 6, 8, 4, 6, 8, 10, 4, 7, 4, 3, 3, 4],
    "D": [6, 5, 12, 4, 8, 5, 10, 8, 6, 9, 5, 4, 6, 3],
    "E": [7, 6, 8, 7, 10, 4, 5, 6, 3, 12, 11, 8, 7, 5]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set color palette (rainbow scale)
colors = sns.color_palette("rainbow", as_cmap=True)

# Convert 'Country' column to index
df_rank = df.set_index("Country")

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_rank, cmap=colors, annot=True, fmt="d", linewidths=0.5, cbar=True)
plt.title("Country Level Sales Rank - Top 5 Drugs")
plt.show()
```
