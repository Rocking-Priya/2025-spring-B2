Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

### **Area Plot vs. Two Line Graphs**  

#### **Advantages of an Area Plot:**

Area plot is much easier to understand and compare trends than two separete line graphs. In addition, area plot is very engaging and appealing beacuse of different colors shaded areas than simple lines. The filled area makes it clear how nitrate and phosphate levels change over time. Therefore area plot helps to impove the better visuallization.


#### **Disadvantages of an Area Plot:**

Area plot has some disadvatages as well. According to me, there are two major disadvantages. One is overlapping the values makes hard to to tell individual values of two different features and another is misinterpretation happens when the filled areas might give a false sense of total quantity rather than showing two separate variables clearly. In these type of senariios line graphs plays important role beacuse Line graph is more clear and precise since each line represents one variable and helps us to compare their trends. And Line graphs do not overlap since lines do not fill space, so they don't hide each other.  


#### **How to Improve the Graphs:**  
- **If using an area plot:** Use different colors with transparency to reduce confusion.  
- **If using line graphs:** Use distinct colors and labels, and maybe a legend for clarity.  
- **If values are too close:** Consider using a **dual-axis graph**, where nitrate and phosphate have separate y-axes to highlight differences better.
